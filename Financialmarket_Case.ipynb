```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

import warnings
warnings.filterwarnings('ignore')
```

```python
# Dataset load
df = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx - Tab_Dados(Financialmarket_Case).xlsx')
df.head(5)
```

```python
df.describe()
```

```python
# Time Series
data = df.set_index('Data')
data.head(5)
```

```python
df.tail(5)
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16, 5))
plt.title('Analysis of the stocks/shares of MAGALU - Fechamento', fontsize=14, loc='left')

plt.plot(data.index, data['Fechamento'])

plt.xlabel('Quotation period')
plt.ylabel('Stock value (R$)');
```

```python
moving_average = df['Fechamento'].rolling(5).mean()
trend_average = df['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16, 5))
plt.title('Analysis of the stocks/shares of MAGALU - Fechamento', fontsize=14, loc='left')

plt.plot(data.index, data['Fechamento'])
plt.plot(data.index, moving_average)
plt.plot(data.index, trend_average)

plt.xlabel('Quotation period')
plt.ylabel('Stock value (R$)');
```

```python
sns.boxplot(data=data, x='Fechamento')
```

```python
# Monthly boxplot]
df['Mes'] = df['Data'].dt.month
plt.figure(figsize=(16, 5))
sns.boxplot(data=df, x='Mes', y='Fechamento');
```

```python
df.groupby(['Mes']).describe()['Fechamento']
```

```python
graph = go.Figure(
    data=[
          go.Candlestick(
              x=data.index,
              open=data['Abertura'],
              high=data['Maior'],
              low=data['Menor'],
              close=data['Fechamento'],
          )
    ]
)

graph.update_layout(xaxis_rangeslider_visible=False)

graph.show()
```
