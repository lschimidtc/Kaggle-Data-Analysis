```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Dataset load
df = pd.read_csv('Salary_Data(HR Case).csv')

# Dataset dimension
df.head(5)
```

```python
# Renaming columns
df.rename( columns={
    'YearsExperience' : 'Xp',
}, inplace=True )
```

```python
# Dimension
df.shape
```

```python
# Null fields
df.isnull().sum()
```

```python
# Null graph
sns.heatmap(df.isnull(), cbar=False);
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Income Analysis')
sns.kdeplot(df['Salary'], color='green');
```

```python
plt.figure(figsize=(10, 6))
plt.title('XP Analysis')
sns.kdeplot(df['Xp'], color='blue');
```

```python
# Distribution
plt.figure(figsize=(10, 5))
sns.boxplot(df['Salary']);
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Salary', y='Xp');
```

```python
plt.figure(figsize=(10, 5))
sns.regplot(data=df, x='Salary', y='Xp');
```

```python
# Correlation
Correlation = df.corr('pearson')
plt.figure( figsize=(10, 5) )
sns.heatmap(Correlation, annot=True, cbar=True);
```

```python
Correlation = df.corr('spearman')
plt.figure(figsize=(10, 5))
sns.heatmap(Correlation, annot=True, cbar=True);
```
